```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data (placeholder)
# If dataset is not yet downloaded, mention in report how it will be added
#url=(https://drive.google.com/file/d/1Ojiw4ZPDUoy5t_tBJPhhMdJdMK7cVPEN/view?usp=sharing)
file_id = "1Ojiw4ZPDUoy5t_tBJPhhMdJdMK7cVPEN "
file_url = f"https://drive.google.com/uc?id={file_id}"

df = pd.read_csv(file_url)
print(df.head())

df.head()

```

```python
# Quick profiling
df.info()
df.describe()
#df.isna().sum()
```

```python
df.isna().sum()
```

```python
#Outlier Detection & Cleaning
#Check for outliers using Z-scores (to detect unusual values):
from #scipy import stats
import numpy as np

z = np.abs(stats.zscore(df[['GHI','DNI','DHI','ModA','ModB','WS','WSgust']]))
df_clean = df[(z < 3).all(axis=1)]
```

```python
#Handling missing values
df_clean.fillna(df_clean.median(numeric_only=True), inplace=True)
```

```python
df_clean['Timestamp'] = pd.to_datetime(df_clean['Timestamp'], errors='coerce')
```

```python
df_clean = df_clean.sort_values('Timestamp')
df_clean = df_clean.dropna(subset=['Timestamp', 'GHI', 'DNI', 'DHI'])
```

```python
#Time-Series Analysis
#Plot GHI, DNI, DHI, and Tamb vs Timestamp.

plt.figure(figsize=(12,6))
plt.plot(df_clean['Timestamp'], df_clean['GHI'], label='GHI')
plt.plot(df_clean['Timestamp'], df_clean['DNI'], label='DNI', alpha=0.7)
plt.plot(df_clean['Timestamp'], df_clean['DHI'], label='DHI', alpha=0.7)
plt.title('Benin Solar Radiation Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Irradiance (W/mÂ²)')
plt.legend()
plt.show()
```

```python
print(df.columns)
```

```python
#Cleaning Impact
df_clean.groupby('Cleaning')[['ModA','ModB']].mean().plot(kind='bar')
plt.title('Effect of Cleaning on Sensor Readings')
plt.show()
```

```python
 
#Correlation & Relationship Analysis
corr = df_clean[['GHI','DNI','DHI','TModA','TModB']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plots
sns.scatterplot(x='RH', y='Tamb', data=df_clean)
sns.scatterplot(x='WS', y='GHI', data=df_clean)
```

```python
# Wind & Distribution Analysis
sns.histplot(df_clean['GHI'], kde=True)
sns.histplot(df_clean['WS'], kde=True)
```

```python
#Temperature & Humidity Analysis
sns.scatterplot(x='RH', y='GHI', hue='Tamb', data=df_clean)
```

```python
#Bubble Chart
plt.scatter(df_clean['Tamb'], df_clean['GHI'], s=df_clean['RH']*2, alpha=0.5)
plt.xlabel('Temperature (Â°C)')
plt.ylabel('GHI (W/mÂ²)')
plt.title('GHI vs Temperature (Bubble size = RH)')
plt.show()
```

```python
df_clean.to_csv("../data/benin_clean.csv", index=False)
print(" Cleaned file saved to ../data/benin_clean.csv")
```

### Summary of Findings (Benin)
- The Benin dataset shows strong solar activity during daylight hours, with clear daily cycles in GHI, DNI, and DHI.  
- Outliers were mainly detected in wind speed (WSgust) and sensor readings (ModB), which were filtered using Z-scores.  
- Missing data in temperature and humidity were minimal and handled via median imputation.  
- Cleaning the data improved consistency between ModA and ModB readings, especially after cleaning operations.  
- GHI and DNI have a high positive correlation, confirming data reliability.  
- Relative humidity (RH) and ambient temperature (Tamb) display an inverse relationship â€” as humidity rises, temperature tends to drop.
